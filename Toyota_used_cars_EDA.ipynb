Using flow control to change numerical data to categorical data

```
# Making categorical data by adding a column Price_Class

for i in range(0,len(df['Price']),1):
    if df['Price'][i]<=8450:
        df['Price_Class'][i]='Low'
    elif df['Price'][i]>11950:
        df['Price_Class'][i] = 'High'
    else:
        df['Price_Class'][i] = 'Medium'
df
```

Converting values to ratios using a function

```
def converted(val):
    return val/12

df['Age_Converted'] = converted(df['Age'])
df['Age_Converted'] = round(df['Age_Converted'],1)
df
```

```
df.insert(12,'KM_Per_Month','')
df
```

```
def c_convert(v1,v2):
    return [v1/12,v2/v1]

df['Age_Converted'],df['KM_Per_Month'] = c_convert(df['Age'],df['KM'])
df
```

### Using crosstab in pandas 

```
# Frequency tables

# Checking relationships

pd.crosstab(cars_data2['FuelType'],'count',dropna=True)
```

Relationship bw 2 columns

```
# Relationship bw 2 columns

pd.crosstab(cars_data2['Automatic'],cars_data2['FuelType'],dropna=True)
```

Joint probability

```
# Joint probability
# Likelihood of 2 independent events happening at the same time


# Values converting to probability

pd.crosstab(cars_data2['Automatic'],cars_data2['FuelType'],dropna=True,normalize=True)
```

Marginal probability

```
# Marginal probability

# Occurence of a single event

pd.crosstab(cars_data2['Automatic'],cars_data2['FuelType'],dropna=True,normalize=True)
```

Conditional probability

```
# Conditional probability

# Prob of b if a has occured

pd.crosstab(cars_data2['Automatic'],cars_data2['FuelType'],dropna=True,normalize='index',margins=True)
```

2 way conditional probability

```
# 2 way talbe - Conditional probability

pd.crosstab(cars_data2['Automatic'],cars_data2['FuelType'],dropna=True,normalize='columns',margins=True)
```

## Imputing missing values using replace

```
# Imputing missing value of the age varibale

cars_data3['Age'].mean()
```

```
cars_data3['Age'].fillna(cars_data3['Age'].mean(),inplace=True)
cars_data3
```

```
cars_data3['KM'].median()
```

```
cars_data3['KM'].fillna(cars_data3['KM'].median(),inplace=True)
cars_data3
```

```
cars_data3['HP'].mean()
```

```
cars_data3['HP'].fillna(cars_data3['HP'].mean(),inplace=True)
cars_data3
```
